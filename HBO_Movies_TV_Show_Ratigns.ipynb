## Data Cleaning/Processing

### HBO Content Cleaning

```python
# Drop genres with less than 50
genre_columns = hbo_content_df.filter(regex='^genres_')
genre_counts = genre_columns.sum()
genres_to_drop = genre_counts[genre_counts < 50].index
clean_hbo_df = hbo_content_df.drop(columns=genres_to_drop)
```

```python
# Drop movies/TV series with no genre
clean_hbo_df = clean_hbo_df.drop(clean_hbo_df[clean_hbo_df.filter(regex='^genres_').sum(axis=1) == 0].index)
```

```python
# Drop all platforms except for hbo and hbo_max
clean_hbo_df = clean_hbo_df.drop(columns=clean_hbo_df.filter(regex='^platforms_').columns.difference(['platforms_hbo', 'platforms_hbo_max']))
```

```python
# Drop decade, imdb_bucket, index
clean_hbo_df.drop(columns=['decade', 'imdb_bucket', 'index'], inplace=True)
```

```python
clean_hbo_df.head()
```

```python
clean_hbo_df.shape[0]
```

```python
clean_hbo_df['rating'].value_counts(dropna=False)
```

```python
# Convert rating column to binary variables
dummies = pd.get_dummies(clean_hbo_df['rating'], prefix='rating')
clean_hbo_df = pd.concat([clean_hbo_df, dummies], axis=1)
```

```python
# Convert type 'TV' to 1 and null to 0
clean_hbo_df['type'] = clean_hbo_df['type'].map({'TV': 1, np.nan: 0})
clean_hbo_df.head()
```

The following clean data has this data modified:

* type column -> 0 = Movie, 1 = TV
* all platforms except hbo and hbo_max removed
* dropped genres with less than 50 occurences
* removed movies/TV shows with no genre

HBO content dataset is best for analysing genres, rating, year, type.

Considerations:

* rating and scores has null values

### HBO MAX Content Cleaning

```python
# Drop genres with less than 50 occurrences
genre_columns = hbo_max_content_df.filter(regex='^genres_')
genre_counts = genre_columns.sum()
genres_to_drop = genre_counts[genre_counts < 50].index
clean_hbo_max_df = hbo_max_content_df.drop(columns=genres_to_drop)

# Drop movies/TV series with no genre
clean_hbo_max_df.drop(clean_hbo_max_df[clean_hbo_max_df.filter(regex='^genres_').sum(axis=1) == 0].index, inplace=True)
```

```python
# Drop platforms with less than 100 occurrences
platform_counts = clean_hbo_max_df.filter(regex='^platforms_').sum()
platforms_to_drop = platform_counts[platform_counts < 100].index
clean_hbo_max_df.drop(columns=platforms_to_drop, inplace=True)

# Drop movies/TV series with no platform
clean_hbo_max_df.drop(clean_hbo_max_df[clean_hbo_max_df.filter(regex='^platforms_').sum(axis=1) == 0].index, inplace=True)
```

```python
# Drop decade, imdb_bucket, index
clean_hbo_max_df.drop(columns=['decade', 'imdb_bucket', 'index'], inplace=True)
```

```python
# Convert rating to binary
dummies = pd.get_dummies(clean_hbo_max_df['rating'], prefix='rating')
clean_hbo_max_df = pd.concat([clean_hbo_max_df, dummies], axis=1)
```

```python
# Convert type 'TV' to 1 and null to 0
clean_hbo_max_df['type'] = clean_hbo_max_df['type'].map({'TV': 1, np.nan: 0})
```

```python
clean_hbo_max_df.head()
```

The above clean data has the following modified:

* type column -> 0 = Movie, 1 = TV
* dropped platforms with less than 100 occurences
* dropped genres with less than 50 occurences
* removed movies/TV shows with no genre or platform

HBO MAX content dataset is best for analysing platforms, genres, rotten, imdb, year, rating, type

Considerations:

* rating and scores still has null values

### Merged Content Cleaning

```python
# Drop genres less than 100
genre_columns = merged_df.filter(regex='^genres_')
genre_counts = genre_columns.sum()
genres_to_drop = genre_counts[genre_counts < 100].index
clean_merged_df = merged_df.drop(columns=genres_to_drop)

# Drop movies/TV series with no genre
clean_merged_df.drop(clean_merged_df[clean_merged_df.filter(regex='^genres_').sum(axis=1) == 0].index, inplace=True)

# Drop platforms with less than 100 occurrences
platform_counts = clean_merged_df.filter(regex='^platforms_').sum()
platforms_to_drop = platform_counts[platform_counts < 100].index
clean_merged_df.drop(columns=platforms_to_drop, inplace=True)

# Drop movies/TV series with no platform
clean_merged_df.drop(clean_merged_df[clean_merged_df.filter(regex='^platforms_').sum(axis=1) == 0].index, inplace=True)

# Drop decade, imdb_bucket, index
clean_merged_df.drop(columns=['decade', 'imdb_bucket', 'index'], inplace=True)

# Convert rating to binary
dummies = pd.get_dummies(clean_merged_df['rating'], prefix='rating')
clean_merged_df = pd.concat([clean_merged_df, dummies], axis=1)

# Convert type 'TV' to 1 and null to 0
clean_merged_df['type'] = clean_merged_df['type'].map({'TV': 1, np.nan: 0})

clean_merged_df.head()
```

```python
# Drop platforms_criterion_channel
clean_merged_df.drop(columns=['platforms_criterion_channel'], inplace=True)
clean_merged_df.head()
```

* type column -> 0 = Movie, 1 = TV
* dropped platforms with less than 100 occurences

* removed movies/TV shows with no genre or platform

Merged content dataset is best for analysing genres, rotten, imdb, year, rating, type.
